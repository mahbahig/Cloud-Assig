```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("books.csv")
dataset.head()
```

```python
dataset.dropna(inplace=True)
dataset.drop_duplicates(inplace=True)
dataset['original_publication_year'] = dataset['original_publication_year'].astype(int)
columns_to_drop = ['book_id', 'goodreads_book_id', 'best_book_id', 'work_id', 'isbn', 'isbn13', 'language_code', 'image_url', 'small_image_url']
dataset.drop(columns=columns_to_drop, inplace=True)
dataset.head()
```

```python
harry_potter_books = dataset[dataset['original_title'].str.contains('harry potter', case=False) | dataset['title'].str.contains('harry potter', case=False)]
harry_potter_books.head()
```

```python
print("Harry Potter Books Average Rating:", harry_potter_books['average_rating'].mean())
```

```python
most_selling_books = harry_potter_books.sort_values(by='ratings_count', ascending=False).head(5)

print("Most Selling Harry Potter Books:")
print(most_selling_books[['title', 'ratings_count']])
```
